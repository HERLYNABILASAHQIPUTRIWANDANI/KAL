### 1. Tentukan luas jajaran genjang dari dua vektor 2 dimensi

Diketahui:

$$
\vec{u} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad \vec{v} = \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$

Dalam 2 dimensi, luas jajaran genjang yang dibentuk dua vektor dapat dihitung dengan determinan matriks:

$$
\text{Luas} = \left| \vec{u} \times \vec{v} \right| = \left| \begin{vmatrix} 1 & 2 \\ 2 & 1 \end{vmatrix} \right| = |(1)(1) - (2)(2)| = |1 - 4| = |-3| = {3}
$$


### 2. Tentukan luas jajaran genjang dari dua vektor 3 dimensi

Diketahui:

$$
\vec{u} = \begin{bmatrix} 2 \\ 0 \\ 0 \end{bmatrix}, \quad \vec{v} = \begin{bmatrix} 0 \\ 3 \\ 0 \end{bmatrix}
$$

Luas jajaran genjang dari dua vektor di ruang 3 dimensi adalah:

$$
\text{Luas} = \left| \vec{u} \times \vec{v} \right|
$$

Hitung cross product:

$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 0 & 0 \\
0 & 3 & 0
\end{vmatrix}
= \mathbf{i}(0\cdot0 - 0\cdot3) - \mathbf{j}(2\cdot0 - 0\cdot0) + \mathbf{k}(2\cdot3 - 0\cdot0)
= \mathbf{0i} - \mathbf{0j} + 6\mathbf{k}
$$

Norma dari hasil cross product:

$$
\|\vec{u} \times \vec{v}\| = \sqrt{0^2 + 0^2 + 6^2} = \sqrt{36} = {6}
$$


### 3. Tentukan luas segitiga dari 3 titik di ruang 3 dimensi

Diketahui:

$$
A = (0, 0, 0), \quad B = (1, 3, -1), \quad C = (2, 1, 1)
$$

Langkah-langkah:

1. Hitung vektor AB dan AC:

   $$
   \vec{AB} = B - A = (1, 3, -1), \quad \vec{AC} = C - A = (2, 1, 1)
   $$

2. Hitung cross product $\vec{AB} \times \vec{AC}$:

$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 3 & -1 \\
2 & 1 & 1
\end{vmatrix}
= \mathbf{i}(3\cdot1 - (-1)\cdot1) - \mathbf{j}(1\cdot1 - (-1)\cdot2) + \mathbf{k}(1\cdot1 - 3\cdot2)
= \mathbf{i}(4) - \mathbf{j}(3) + \mathbf{k}(-5)
= (4, -3, -5)
$$

3. Norma dari vektor hasil cross product:

$$
\| \vec{AB} \times \vec{AC} \| = \sqrt{4^2 + (-3)^2 + (-5)^2} = \sqrt{16 + 9 + 25} = \sqrt{50}
$$

4. Luas segitiga adalah:

$$
\text{Luas} = \frac{1}{2} \cdot \| \vec{AB} \times \vec{AC} \| = \frac{1}{2} \cdot \sqrt{50} = \frac{1}{2} \cdot 5\sqrt{2} = {\frac{5\sqrt{2}}{1} \approx 3.54}
$$


### 4. Tentukan luas segitiga dari 3 titik di ruang 3 dimensi

Diketahui:

$$
A = (5, 2, -1), \quad B = (3, 6, 2), \quad C = (1, 0, 4)
$$

Langkah-langkah:

1. Hitung vektor AB dan AC:

   $$
   \vec{AB} = B - A = (-2, 4, 3), \quad \vec{AC} = C - A = (-4, -2, 5)
   $$

2. Hitung cross product $\vec{AB} \times \vec{AC}$:

$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
-2 & 4 & 3 \\
-4 & -2 & 5
\end{vmatrix}
= \mathbf{i}(4\cdot5 - 3\cdot(-2)) - \mathbf{j}((-2)\cdot5 - 3\cdot(-4)) + \mathbf{k}((-2)\cdot(-2) - 4\cdot(-4))
$$

$$
= \mathbf{i}(20 + 6) - \mathbf{j}(-10 + 12) + \mathbf{k}(4 + 16) = (26, -2, 20)
$$

3. Norma dari vektor hasil cross product:

$$
\| \vec{AB} \times \vec{AC} \| = \sqrt{26^2 + (-2)^2 + 20^2} = \sqrt{676 + 4 + 400} = \sqrt{1080}
= \sqrt{36 \cdot 30} = 6\sqrt{30}
$$

4. Luas segitiga:

$$
\text{Luas} = \frac{1}{2} \cdot \| \vec{AB} \times \vec{AC} \| = \frac{1}{2} \cdot 6\sqrt{30} = {3\sqrt{30} \approx 16.43}
$$
